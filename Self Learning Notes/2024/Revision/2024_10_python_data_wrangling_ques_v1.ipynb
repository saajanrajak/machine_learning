In pandas, whether to use `in` or `.str.contains()` depends on what you are trying to achieve, particularly with string matching and filtering. Here’s a breakdown of when to use each:

### 1. Using `in`

- **Purpose**: `in` is used to check if a substring exists within a string.
- **Context**: Typically used in Python for checking membership, such as when you want to verify if a specific value is present in a list or string.
- **Example**: When you are working with individual strings, you can use `in` to see if a substring exists.

#### Example
```python
# Example of using 'in' for string checking
name = "Mr. John Doe"
if "Mr." in name:
    print("Title found!")
```

### 2. Using `.str.contains()`

- **Purpose**: `.str.contains()` is a pandas method that checks if a pattern or substring is present in each element of a Series.
- **Context**: Use this when you are working with a pandas Series and want to perform vectorized string matching. It supports regular expressions and can handle NaN values gracefully.
- **Example**: When filtering a DataFrame based on a condition that involves substrings within a column of strings.

#### Example
```python
import pandas as pd

# Example DataFrame
data = {'Name': ['Mr. John Doe', 'Mrs. Jane Smith', 'Miss Emily Davis']}
df = pd.DataFrame(data)

# Filtering using .str.contains()
filtered_df = df[df['Name'].str.contains('Mrs.')]
print(filtered_df)
```

### Key Differences

1. **Scope**:
   - `in`: Works with individual strings (or lists).
   - `.str.contains()`: Works with pandas Series and can handle each element of the Series.

2. **Use Case**:
   - Use `in` for simple checks within a string or list.
   - Use `.str.contains()` for filtering DataFrame rows based on the presence of a substring in a Series.

3. **Regular Expressions**:
   - `.str.contains()` can accept regular expressions, allowing for more complex matching patterns.

### Conclusion

- **Use `in`** when checking membership in single strings or for simple substring checks.
- **Use `.str.contains()`** when working with pandas Series for filtering and more complex string matching, especially when you want to leverage regular expressions or handle multiple elements efficiently.

- When you use df['Embarked'].mode(), it returns a Series (as there could be multiple modes if there is a tie), but .fillna() expects a single value.
- while df['age'].median() gives single value
- The difference between `.mode()` and `.median()` lies in how these methods handle their outputs based on the nature of the statistics they calculate:

1. **`.mode()`**:
   - **Purpose**: Finds the most frequently occurring value(s) in a Series.
   - **Output**: Returns a **Series**, as there may be multiple modes if there’s a tie.
   - **Example**: If two values appear with the same highest frequency, `.mode()` will return both as a Series. Even if there's only one mode, it still returns a Series for consistency.

2. **`.median()`**:
   - **Purpose**: Calculates the middle value (or the average of the two middle values) in a sorted Series.
   - **Output**: Returns a **single value** (a scalar), as there can only be one median value.
   - **Example**: Regardless of the data distribution, the median is always a unique central value, so `.median()` provides a single scalar.

This is why `.mode()` returns a Series, while `.median()` returns a single value.

````python
def extract_deck_mode(cabin_series):
    # Step 1: Handle NaN values and extract the first letter of each cabin entry
    decks = cabin_series.dropna().str.split(' ').apply(lambda x: [cabin[0] for cabin in x if isinstance(cabin, str)])
    
    # Step 2: Find the most common deck level for each row, accounting for NaN values
    deck_mode = decks.apply(lambda x: pd.Series(x).mode()[0] if len(x) > 0 else None)
    
    # Fill NaN values for rows that were NaN originally
    return deck_mode.reindex(cabin_series.index)

# Apply the function to the 'Cabin' column
df['Deck'] = extract_deck_mode(df['Cabin'])
df
````

Yes, there are several alternative methods to create a unique ID for each family group based on `Surname` and `Pclass` in a pandas DataFrame. Below are a few different approaches, each with its own advantages.

### Alternative Method 1: Using `factorize()`

You can use the `factorize()` function to create unique IDs directly from the unique combinations of `surname` and `Pclass`. This method is concise and efficient.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Smith, John', 'Doe, Jane', 'Smith, Alice', 'Doe, Jim', 'Brown, Charlie', 'Brown, David'],
    'Pclass': [1, 1, 2, 2, 3, 3]
}

df = pd.DataFrame(data)

# Extract the Surname
df['surname'] = df["Name"].str.split(', ').str[0]

# Create a unique identifier using factorize
df['surname_pclass_id'] = pd.factorize(df['surname'].astype(str) + '_' + df['Pclass'].astype(str))[0] + 1

# Display the updated DataFrame
print(df)
```

### Output

```
             Name  Pclass   surname  surname_pclass_id
0   Smith, John      1     Smith                  1
1   Doe, Jane       1      Doe                    2
2  Smith, Alice     2     Smith                  1
3   Doe, Jim        2      Doe                    2
4  Brown, Charlie   3    Brown                  3
5   Brown, David    3    Brown                  3
```

### Explanation

- **`pd.factorize()`**: This function encodes the unique combinations of the `surname` and `Pclass` as integer codes. The resulting IDs start from 0, so we add 1 to start from 1.

### Alternative Method 2: Using `groupby()` with `ngroup()`

Another approach is to use `groupby()` combined with `ngroup()`, which assigns a group number to each unique combination of `surname` and `Pclass`.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Smith, John', 'Doe, Jane', 'Smith, Alice', 'Doe, Jim', 'Brown, Charlie', 'Brown, David'],
    'Pclass': [1, 1, 2, 2, 3, 3]
}

df = pd.DataFrame(data)

# Extract the Surname
df['surname'] = df["Name"].str.split(', ').str[0]

# Create a unique identifier using groupby and ngroup
df['surname_pclass_id'] = df.groupby(['surname', 'Pclass']).ngroup() + 1

# Display the updated DataFrame
print(df)
```

### Output

```
             Name  Pclass   surname  surname_pclass_id
0   Smith, John      1     Smith                  1
1   Doe, Jane       1      Doe                    2
2  Smith, Alice     2     Smith                  1
3   Doe, Jim        2      Doe                    2
4  Brown, Charlie   3    Brown                  3
5   Brown, David    3    Brown                  3
```

### Explanation

- **`groupby(['surname', 'Pclass'])`**: This groups the DataFrame by the `surname` and `Pclass` columns.
- **`ngroup()`**: This assigns a unique number to each group starting from 0, so we add 1 to start from 1.

### Alternative Method 3: Using `apply()` with Custom Function

If you prefer a more custom approach, you can use `apply()` to create a unique ID by concatenating `surname` and `Pclass` and then using `factorize()`.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Smith, John', 'Doe, Jane', 'Smith, Alice', 'Doe, Jim', 'Brown, Charlie', 'Brown, David'],
    'Pclass': [1, 1, 2, 2, 3, 3]
}

df = pd.DataFrame(data)

# Extract the Surname
df['surname'] = df["Name"].str.split(', ').str[0]

# Create a unique identifier using apply and factorize
df['surname_pclass_id'] = df.apply(lambda x: f"{x['surname']}_{x['Pclass']}", axis=1)
df['surname_pclass_id'] = pd.factorize(df['surname_pclass_id'])[0] + 1

# Display the updated DataFrame
print(df)
```

### Output

```
             Name  Pclass   surname  surname_pclass_id
0   Smith, John      1     Smith                  1
1   Doe, Jane       1      Doe                    2
2  Smith, Alice     2     Smith                  1
3   Doe, Jim        2      Doe                    2
4  Brown, Charlie   3    Brown                  3
5   Brown, David    3    Brown                  3
```

### Summary

- **Using `factorize()`**: Quick and efficient for unique IDs based on combined columns.
- **Using `groupby()` with `ngroup()`**: Assigns group numbers easily for unique combinations.
- **Using `apply()` with a custom function**: Provides flexibility for more complex conditions.
